```python
# importing numpy library
import numpy as np
```

```python
# creating and reshaping Arrays
# 1.create arrays
# A 1D array of 10 equally spaced values between 0 and 50.
arr1 = np.linspace(0,50,10)
arr1
```

```python
# A 2D array of shape (3x3) containing random integers between 5 and 20.
arr = np.random.randint(5,20,(3,3))
arr.reshape(3,3)
```

```python
# 2)print dimensions of both arrays
arr.ndim

```

```python
arr1.ndim
```

```python
# Print the shape of both arrays.
arr.shape

```

```python
arr1.shape
```

```python
# 3) reshape the 1D array into 2D
# Reshape the 1D array into a 2D array with shape (2x5).
# print the reshaped array
arr1 = arr1.reshape(2,5)
arr1
```

```python
# Flatten the reshaped array back to 1D and print it.
arr1 = arr1.reshape(-1)
arr1.shape
```

```python
#part - 5
# Write a program that asks the user to input a number (n), and then
# generates an n x n array of random integers between 1 and 100


n = int(input("Enter n: "))
array = np.random.randint(1,101,(n,n))
print(array)
print(array.shape)
```
