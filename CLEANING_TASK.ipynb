```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/Taj1920/Data-Analysis-Datasets/raw/refs/heads/main/Datasets-Cleaning/USA_presidents.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.dropna(inplace=True) #removing null values
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# removing unwanted columns
df = df.drop("Unnamed: 0.1",axis=1)
```

```python
df = df.drop("Unnamed: 0",axis=1)
```

```python
# removing duplicate rows
df.shape #42
df.drop_duplicates(inplace=True)
df.shape #41
```

```python
df
```

```python
# df.info()
df["S.No."].unique()
```

```python
# handling s.no. column
df["S.No."] = range(1, len(df) + 1)
```

```python
df["S.No."].unique()
```

```python
df.info()
```

```python
df["president"].unique()
df["president"] = df["president"].str.title()
```

```python
df["prior"].unique()
```

```python
df["party"].unique()
```

```python
# handling party column
df["party"] = df["party"].replace({
    "Demorcatic": "Democratic",
    "Republicans": "Republican"
})
```

```python
df["party"] = df["party"].str.replace(
    "Whig   April 4, 1841  â€“  September 13, 1841",
    "Whig"
)
```

```python
df["party"] = df["party"].replace(
    "Democratic-  Republican",
    "Democratic-Republican"
)
```

```python
df["party"].unique()
```

```python
df["vice"].unique()
```

```python
df.rename(columns={"vice": "vice_president"}, inplace=True)
```

```python
df["vice_president"].str.strip()
df["vice_president"] = df["vice_president"].str.split().str.join(" ")
df["vice_president"].unique()
```

```python
df["salary"].unique()
```

```python
# handling salary column
df["salary"] = df["salary"].str.replace("$", "")
df["salary"] = df["salary"].str.replace(",", "")
df["salary"] = df["salary"].astype(int)
df["salary"].unique()
```

```python
df["date updated"].unique()
```

```python
df["date created"].unique()
```

```python
df.reset_index(inplace=True)
df = df.drop("index",axis=1)
df
```
