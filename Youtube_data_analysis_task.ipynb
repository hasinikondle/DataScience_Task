```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("https://github.com/Taj1920/Data_Analysis/raw/refs/heads/main/Youtube_analysis/YOUTUBE%20CHANNELS%20DATASET.csv")
```

# Data Cleaning

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# handling null values
# df["Ranking"] = df["Ranking"].fillna("0")
df["Ranking"] = range(1, len(df) + 1)
df["Ranking"].unique()
```

```python
df["Username"] = df["Username"].str.capitalize()
mode = df["Username"].mode()
df["Username"] = df["Username"].fillna("Zhong")
df["Username"].unique()
```

```python
df["Subscribers"].unique()
mode = df["Subscribers"].mode()
df["Subscribers"] = df["Subscribers"].fillna("62.9M")
df["Subscribers"].unique()
df["Subscribers"] = df["Subscribers"].str.replace("M","")
df["Subscribers"].unique()
df["Subscribers"]=df["Subscribers"].astype(float)
```

```python
df["Uploads"] = df["Uploads"].str.replace(",","")
```

```python
mode = df["Uploads"].mode()
df["Uploads"] = df["Uploads"].fillna(0)
```

```python
df["Uploads"] = df["Uploads"].astype(int)
```

```python
df["Views"] = df["Views"].str.replace(",","")
```

```python
mode = df["Views"].mode()
df["Views"] = df["Views"].fillna(0)
```

```python
df["Views"] = df["Views"].astype(int)
```

```python
mode = df["Country"].mode()
df["Country"] = df["Country"].fillna("IN")
```

```python
data = df[["Subscribers", "Uploads", "Views"]].corr()
sns.heatmap(data,annot=True)
```

1)**Uploads do NOT determine subscribers**

Correlation = 0.0039

Almost no relationship.

Uploading more videos does NOT guarantee more subscribers.

2)**Subscribers strongly relate to Views**

Correlation = 0.63

More subscribers → more total views

3)**Uploads have weak relation with Views**

Correlation = 0.18

Quantity strategy is not very powerful.

Quality seems more important.


# Insights :

1)Uploading more videos does not strongly increase subscribers.



2)Subscriber count has a strong influence on total views.



3)Entertainment creators achieve high engagement with fewer uploads.



4)News & TV channels rely on bulk upload strategy.

# Recommendations:

1)Focus on content quality rather than upload frequency.

2)Build subscriber base for sustainable long-term views.

3)Use high-volume strategy only if content production cost is low.

4)Optimize per-video impact instead of chasing quantity.
